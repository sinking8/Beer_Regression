```python
#removing id's and names
df = df.drop(['index','beer/beerId','user/profileName','beer/brewerId'], axis = 1)

##Removing below columns
#user/ageInSeconds    29644
#user/birthdayRaw     29644
#user/birthdayUnix    29644
#user/gender          22186
df = df.drop(['user/ageInSeconds','user/birthdayRaw','user/birthdayUnix','user/gender'],axis  =1)

#Removing review/text,review/timeStruct,review/timeUnix
df = df.drop(['review/text','review/timeStruct','review/timeUnix'], axis = 1)
```

```python
y  = df['review/overall']
x  = df.drop(['review/overall'], axis = 1)
```

## Scaling and Train Test Split

### Scaling

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

## Encoding for beer/name
le_beer_name   = LabelEncoder()
x['beer/name'] = le_beer_name.fit_transform(x['beer/name'])

## Encoding for beer/style
le_beer_style   = LabelEncoder()
x['beer/style'] =le_beer_style.fit_transform(x['beer/style'])

## Scaling beer/ABV
scaler = MinMaxScaler()
beer_ABV = x['beer/ABV'].values
beer_ABV = beer_ABV.reshape(37500,1)
x['beer/ABV'] = scaler.fit_transform(beer_ABV)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size = 0.20,random_state = 101)
```

```python
print("Shape of y_test  : ",y_test.shape)
print("Shape of X_train : ",X_train.shape)
```

```python
print(x.head(5))
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train,y_train)

preds = np.round(linear_model.predict(X_test),1)
```

### Validation Metrics


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE : ",np.round(mean_absolute_error(y_test,preds),2))
print("MSE : ",np.round(mean_squared_error(y_test,preds),2))
print("R2  : ",np.round(r2_score(y_test,preds),2))
```

## Neural Networks

### Neural Network Model

```python
import tensorflow as tf

model =  tf.keras.Sequential()


model.add(tf.keras.layers.Dense(7,activation='relu'))
model.add(tf.keras.layers.Dense(7,activation='relu'))
model.add(tf.keras.layers.Dense(7,activation='relu'))

model.add(tf.keras.layers.Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train.values,
          validation_data=(X_test,y_test.values),epochs=100)
```

```python
preds = np.round(model.predict(X_test),2)
```

### Validation Metrics

```python
print("MAE : ",np.round(mean_absolute_error(y_test,preds),2))
print("MSE : ",np.round(mean_squared_error(y_test,preds),2))
print("R2  : ",np.round(r2_score(y_test,preds),2))
```
